```python
import cv2
import numpy as np
import insightface
from insightface.app import FaceAnalysis
import matplotlib.pyplot as plt
```

```python
app = FaceAnalysis()
app.prepare(ctx_id=0, det_size=(640, 640))
```

```python
image_file = "./images/joji1.png"

img = cv2.imread(image_file)
temp = img[:,:,::-1]

plt.imshow(temp,cmap='gray')
plt.axis('off')
plt.show()
```

```python
faces = app.get(img)
print(faces)
rimg = app.draw_on(img, faces)
temp = rimg[:,:,::-1]
plt.imshow(temp,cmap='gray')
plt.axis('off')
plt.show()

# cv2.imwrite("./output_images/output.jpg", rimg)
```
